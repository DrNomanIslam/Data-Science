```python
import pandas as pd
```

```python
df = pd.read_csv('company_sales_data.csv')
```

```python
df
```

```python
#Total profit data provided for each month
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(df['month_number'],df['total_profit'], color='k')
plt.xlabel('Month')
plt.ylabel('Total Profit')
```

```python
#customizing plots
```

```python
plt.plot(df['month_number'],df['total_profit'], 
         marker='o', linestyle='dashed' ,linewidth=3,color='red', label='Total Profit')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.legend(loc='lower right')
```

```python
#Products Vs Total Sale
for i in range(2,8):
    plt.plot(df['month_number'],df.iloc[:,i], label = df.columns[i])
plt.legend()
```

```python
#Only toothpase data with grid
```

```python
plt.scatter(df['month_number'], df['toothpaste'])
plt.grid(which='major',linestyle='dashed')
plt.xlabel('Month')
plt.ylabel('Toothpaste')
```

```python
#barchart
```

```python
plt.bar(df['month_number']-0.25,df['facewash'], width=0.25,label='Facewash')
plt.bar(df['month_number']+0.25,df['facecream'], width=0.25,label='Facecream')
plt.legend(loc='upper right')
plt.xticks(df['month_number'].tolist())
```

```python
#plot soap data
plt.bar(df['month_number'],df['bathingsoap'])
plt.title('Bathing Soap Sales data')
plt.grid(linestyle='dashed')
```

```python
plt.hist(df['total_profit']/1000,[150, 175, 200, 225, 250, 300, 350])
```

```python
#piechart
```

```python
plt.pie(df['bathingsoap'],labels=df['month_number'],autopct="%1.1f%%")
plt.legend()
```

```python
#Products Vs Total Sale Stackplot
plt.stackplot(df['month_number'],df.iloc[:,1:7].T, labels=df.columns[1:7])
plt.legend()
```
